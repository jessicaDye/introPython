### Python Socket Module

to pull-in socket functionality we use the Python **import** keyword
    import socket
## 4 Layers of the TCP/IP model are:

*Application

*Transport

*Internetworking

*Link

The socket module provides an application an interface for talking to the transport layer.

**Object Oriented Programming** OOP, combining data and functions into objects (often called classes)

functions that are associated with an object get a special name **method**

data(variables) associated with a class object **attribute**

**object** - in OOP, has methods and attrtibutes.

```python
#Send a TCP Packet
import socket #pull in socket functionality

#create socket object
s = socket.socket()


#Prepare a Message
msg = "TCP makes a connection that is reliable and slow! Jessica"

#Convert to binary
m = msg.encode()

#make a connection
s.connect(("35.192.97.212", 12001)) #TCP change from 12000 to 12001

#send the 1's and 0's
s.sendall(m)

print("Packet sent!")

s.close()
```
